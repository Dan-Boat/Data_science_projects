# Stacking of Deep Learning Models (Ensemble model)

### This notebook demonstrate how to use stacking ensemble on different neural networks. The model adopts sklearn database to tackle a multi-class classification problem. 
#### Since the aim is to demonstrate this approach, a simple Multi-Layer Perceptron model using tensorflow is used. However, the approach can be used for complex dense networks when complicated problems are addressed. We optimize the model with Adam flavour of stochastic gradient descent

### Stacked Generalization ensemble, training and saving models, integrated and separate stacking mode (with logistic regression)

```python
# importing models 

from sklearn.datasets import make_blobs 
import matplotlib.pyplot as plt 
import numpy as np
import tensorflow as tf
import pandas as pd  
from keras.models import Sequential 
from keras.layers import Dense 
from os import makedirs
from sklearn.metrics import accuracy_score 
from sklearn.linear_model import LogisticRegression 
from keras.models import load_model
from numpy import dstack
from keras.layers import Input, Dense
from keras.models import Model
from tensorflow.keras.utils import plot_model
from keras.layers.merge import concatenate
from tensorflow.keras.utils import to_categorical 
```

```python
# data preparation and visualizaiton 
X,y = make_blobs(n_samples=1000, centers=3, n_features=2, cluster_std=2, random_state=2)
df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:"red", 1:"black", 2:"blue"}
plt.style.use("seaborn")
plt.rcParams.update({"font.size":22, "font.weight":"bold"})
fig,ax = plt.subplots()

gp = df.groupby("label")
for key, group in gp:
    group.plot(ax=ax, kind="scatter", x="x", y="y", label=key, color=colors[key])


```

```python
X,y = make_blobs(n_samples=1100, centers=3, n_features=2, cluster_std=2, random_state=2)

# one hot encode of y bcos its a multi classfication problem (3 outputs)
y = tf.keras.utils.to_categorical(y)

#split data
n_train = 100
X_train, X_test = X[:n_train, :], X[n_train:,:]
y_train, y_test =  y[:n_train], y[n_train:]
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# funciton to defined network and fitting 
def fit_model(X,y):
    model = Sequential()
    model.add(Dense(25, input_dim=2, activation="relu"))
    model.add(Dense(3, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    #fit model
    model.fit(X, y, epochs=500, verbose=0)
    return model
```

```python
# make directory to store models
makedirs("models")
```

```python
n_members = 5
for i in range(n_members):
    model= fit_model(X=X_train, y=y_train)
    filename = "models/model_" + str(i + 1) + ".h5"
    model.save(filename)
    print("> Saved %s" % filename)
```

```python
# data preparation 
X,y = make_blobs(n_samples=1100, centers=3, n_features=2, cluster_std=2, random_state=2)

# one hot encode of y bcos its a multi classfication problem (3 outputs)
#y = tf.keras.utils.to_categorical(y)

#split data
n_train = 100
X_train, X_test = X[:n_train, :], X[n_train:,:]
y_train, y_test =  y[:n_train], y[n_train:]
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# load models
def load_all_models(n_models):
    all_models = []
    for i in range(n_models):
        filename = "models/model_" + str(i + 1) + ".h5"
        model = load_model(filename)
        all_models.append(model)
        print(">loaded %s" % filename)
        
    return all_models
n_members = 5
members = load_all_models(n_members)
print("Loaded %d models" % len(members))
```

```python
for model in members:
    y_test_enc = tf.keras.utils.to_categorical(y_test)
    _, acc = model.evaluate(X_test, y_test_enc)
```

```python
#create stacked model input dataset as outputs from the ensemble 
def stacked_datasets(members, inputX):
    stackX = None 
    for model in members:
        #make prediction
        yhat = model.predict(inputX, verbose=0)
        #stack into rows, members * probabilities
        if stackX is None:
            stackX = yhat
        else:
            stackX = dstack((stackX, yhat))
    stackX = stackX.reshape((stackX.shape[0], stackX.shape[1]*stackX.shape[2]))
    
    return stackX

def fit_stacked_model(members, inputX, inputy):
    stackedX = stacked_datasets(members, inputX)
    model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
    print(stackedX.shape)
    model.fit(stackedX, inputy)
    return model 
def stacked_prediction(members, model, inputX):
    stackedX = stacked_datasets(members, inputX)
    yhat = model.predict(stackedX)
    return yhat 
```

```python
model_stack = fit_stacked_model(members, X_test, y_test)
yhat = stacked_prediction(members, model_stack, X_test)
acc = accuracy_score(y_test, yhat)
print("Stacked Test Accuracy: %.3f" % acc)
```

```python
# using integrated stacking model 
# update all models to not trainable 
for i in range(len(members)):
    model = members[i]
    for layer in model.layers:
            layer.trainable = False
            
            # rename 
            layer._name = "ensemble_" + str(i+1) + "_" + layer.name
```

```python
def defined_stacked_model(members):
    # update all models to not trainable 
    for i in range(len(members)):
        model = members[i]
        for layer in model.layers:
                layer.trainable = False

                # rename 
                layer._name = "ensemble_" + str(i+1) + "_" + layer.name

    #defined multi-headed output 
    ensemble_visible = [model.input for model in members]

    # merge output from each model 
    ensemble_outputs = [model.output for model in members]

    merge = concatenate(ensemble_outputs)
    hidden = Dense(10, activation="relu")(merge)
    output = Dense(3, activation="softmax")(hidden)
    model = Model(inputs=ensemble_visible, outputs=output)
    
    #plot
    plot_model(model, show_shapes=True, to_file="model_graph.png")
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    
    return model

def fit_stacked_model(model, X, y):
    X = [X for _ in range(len(model.input))]
    y = to_categorical(y)
    
    model.fit(X, y, epochs=300, verbose=0)
    
def predict_stacked_model(model, X):
    X = [X for _ in range(len(model.input))]
    
    return model.predict(X, verbose=0)
    
```
